```python
df['market_value_in_eur'].describe()
```

```python
compwise_data = df.groupby(['player_club_domestic_competition_id'], as_index=False)['market_value_in_eur'].mean().sort_values(by='market_value_in_eur', ascending=False)
print(compwise_data)
```

```python
sb.set(rc={'figure.figsize':(15,5)})
sb.barplot(data = compwise_data, hue= 'player_club_domestic_competition_id',y= 'market_value_in_eur')
```

```python
# ---> It is observed that  players from club competion:GB1 have significantly greater market valuation on average 
#      compared to players from other club competitions.
```

```python
comp_data = df.groupby('player_club_domestic_competition_id').size()
mt.figure(figsize=(10, 30))
sb.displot(data= comp_data,kde=True)
```

```python
comp_data = df.groupby('player_club_domestic_competition_id').size()
print(comp_data)
```

```python
#  ---> The competition sample data is approximately normal (slightly right skewed) , 
##      therefore the inference about market valuation based on competition is consistent. 

# ==> Here I analysed the competition population sample distribution, checked for normality
##    and experimented using 'distplot' for plotting the histogram setting 'kde' to true
###   for Kernel Density Estimation to get an idea of the normality of the sample. 
```

```python
clubwise_data = df.groupby(['current_club_id'], as_index=False)['market_value_in_eur'].sum().sort_values(by='market_value_in_eur', ascending=False)
clubwise_data.head(20)
```

```python
# ---> Club with club_id 583 has the most valuable set of players i.e. most valuable team in the dataset followed
##     by club_id 131 and 985.

#==> Here I recognized the club with highest net market value using the sum function.
```

```python
best_data = df.groupby(['current_club_id','player_id'], as_index=False)['market_value_in_eur'].max().sort_values(by='market_value_in_eur', ascending=False)
print(best_data)
```
